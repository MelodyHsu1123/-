```python
# ch3_24.py
dist = 384400                    # 地球到月亮距離
speed = 1225                     # 馬赫速度每小時1225公里
total_hours = dist // speed      # 計算小時數
days = total_hours // 24         # 商 = 計算天數
hours = total_hours % 24         # 餘數 = 計算小時數
print("總共需要天數")
print(days)
print("小時數")
print(hours)
```

# 4-7-1:設計攝氏溫度和華氏溫度的轉換
### 攝氏溫度 =(華氏溫度-32)*5/9
### 華氏溫度 =攝氏溫度*(9/5)+32

### 程式實例ch4_20.py:請輸入華氏溫度，這個程式會輸出攝氏溫度。

```python
# ch4_20.py
f = float(input("請輸入華氏溫度:"))
c = ( f - 32 ) * 5 / 9
print("華氏 %4.1f 等於攝氏 %4.1f" % (f,c))
```

# 練習:輸入攝氏溫度，輸出華氏溫度。

```python
c = float(input("請輸入攝氏溫度:"))
f =  (c) * ( 9 / 5 ) + 32 
print("攝氏 %4.1f 等於華氏 %4.1f" % (c,f))
```

# 程式實例ch5_4.py:奇數偶數的判斷。

```python
# ch5_4.py
print("奇數偶數判斷")
num = input("請輸入任意整數值")
rem = int(num) % 2
if (rem == 0):
    print("%d 是偶數" % int(num)) 
else:
    print("%d 是奇數" % int(num))
```
