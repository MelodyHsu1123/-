```python
print('計算最終成績')
score=input('請輸入分數')
sc=int(score)
if(sc>=90):
    print('A')
elif(sc>=80):
    print('B')
elif(sc>=70):
    print('C')
elif(sc>=60):
    print('D')
else:
    print('F')
```

```python
print('計算生肖')
year=input('請輸入出生西元年分')
yr=int(year)
yr=yr-1900
rem=yr%12
if(rem==0):
    print('鼠')
elif(rem==1):
    print('牛')
elif(rem==2):
    print('虎')
elif(rem==3):
    print('兔')
elif(rem==4):
    print('龍')
elif(rem==5):
    print('蛇')
elif(rem==6):
    print('馬')
elif(rem==7):
    print('羊')
elif(rem==8):
    print('猴')
elif(rem==9):
    print('雞')
elif(rem==10):
    print('狗')
else:
    print('豬')
```

```python
print('計算票價')
age=int(input('請輸入年齡'))
price=100
if(age<=6 or age>=80):
    fin=price*0.2
    print(fin,'元')
elif(7<=age<=12 or 60<=age<=79):
    fin=price*0.5
    print(fin,'元')
else:
   print('100元')
```
