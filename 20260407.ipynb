### 程式實例 ch5_5.py： 請輸入數字分數，程式將回應A、B、C、D或F等級。

```python
# ch5_5.py
print("計算最終成績")
score = input("請輸入分數: ")
sc = int(score)
if (sc >=90):
    print(" A")
elif (sc >=80):
    print(" B")
elif (sc >=70):
    print(" C")
elif (sc >=60):
    print(" D")
else:
    print(" F")
```

## 練習5-1：出生年 year 先減去 1900，再計算 rem = year%12。rem==0 是屬鼠,rem=1 是屬牛, ...rem==11是屬豬.請寫一程式，可輸入西元年，輸出生肖。


```python
print("計算生肖")
year = input("請輸入西元出生年份: ")
yr=int(year)
yr=yr-1900
rem=yr%12
if(rem==0):
    print(" 鼠")
if(rem==1):
    print(" 牛")
if(rem==2):  
    print(" 虎")
if(rem==3):
    print(" 兔")
if(rem==4):
    print(" 龙")
if(rem==5):
    print(" 蛇")
if(rem==6):
    print(" 马")
if(rem==7):
    print(" 羊")
if(rem==8):  
    print(" 猴")
if(rem==9):
    print(" 鸡")
if(rem==10):
    print(" 狗")
if(rem==11):
    print(" 豬")
```
